## Visualization
Run the the 2 scripts below to have a look of the distribution of all our preferences. (p.s. for the VIPs)


```
import matplotlib.pyplot as plt
import numpy as np

# all preferences visualization
tb_value = list(choice_vip_dict.values())
bins_N = np.unique(tb_value).shape[0]
plt.hist(tb_value, bins=bins_N, histtype='barstacked', rwidth=0.5, range=(0, bins_N), align='left', label=choice_vip_dict.keys())
plt.title('The VIPs\' all 3 selections')
plt.tick_params(labelrotation=90)
plt.legend()
plt.show()
```

1st, 2nd and 3rd preferences visualization.

```
# 1st, 2nd amd 3rd preferences visualizaton
for i in range(3):
    np_tb_value = np.array(tb_value)[:, i, np.newaxis].tolist()
    bins_Ni = np.unique(np_tb_value).shape[0]
    plt.hist(np.transpose(np_tb_value), bins=bins_Ni, histtype='barstacked', rwidth=0.5, range=(0, bins_Ni), align='left', label=choice_vip_dict.keys())
    plt.tick_params(labelrotation=90)
    plt.title('The VIPs\' {} selection'.format(i+1))
    plt.legend()
    plt.show()
```
